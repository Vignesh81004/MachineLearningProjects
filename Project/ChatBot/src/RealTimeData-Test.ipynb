```python
import json
with open("TestData.json", "r") as D:
    Data = json.load(D)
Data
```

```python
XY = []
for i in Data["Data"]:
    XY.append([i["question"],i["tag"]])
```

```python
import random

random.shuffle(XY)
random.shuffle(XY)
```

```python
%run Model.ipynb import NeuralNet
%run NLTK.ipynb import Tokenize,BagOfWord
from sklearn.metrics import classification_report

FILE = "Model2.pth"
data = torch.load(FILE)

input_size = data["input_size"]
hidden_size = data["hidden_size"]
output_size = data["output_size"]
all_words = data['all_words']
tags = data['tags']
model_state = data["model_state"]
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

```python
model = NeuralNet(input_size, hidden_size, output_size).to(device)
model.load_state_dict(model_state)
model.eval()
```

```python
Correct = 0
Wrong = 0
TP = 0  
FP = 0  
FN = 0  

for pattern, Y in XY:
    Words = Tokenize(pattern)
    X = BagOfWord(Words, all_words)
    
    X = X.reshape(1, X.shape[0])
    X = torch.from_numpy(X).to(device)

    output = model(X)
    _, predicted = torch.max(output, dim=1)

    tag = tags[predicted.item()]

    if Y.lower() == tag.lower():
        Correct += 1
        TP += 1  
    else:
        Wrong += 1
        FN += 1  
        FP += 1  
        print(f"Pattern : {pattern}")
        print(f"Bot = {tag}, Dataset = {Y}")


if TP + FP > 0:
    precision = TP / (TP + FP)
else:
    precision = 0.0  
if TP + FN > 0:
    recall = TP / (TP + FN)
else:
    recall = 0.0  
if precision + recall > 0:
    f1_score = 2 * (precision * recall) / (precision + recall)
else:
    f1_score = 0.0  

print("Correct : ", Correct)
print("Wrong : ", Wrong)
print(f"Intent Classification Accuracy : {Correct/207:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1_score:.2f}")
```
